# Embedding Word2Vec


```python
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet', 'por')
nltk.download('wordnet')
nltk.download('omw')
```

```python
import nlpaug.flow as naf
aug_ppdb = naw.SynonymAug(aug_src='ppdb', model_path='/content/drive/My Drive/Colab Notebooks/parafraseadores/ppdb-2.0-m-all')
```

```python
exp_ppdb = []

for i in range(repeticoes):
  base_cru_1p_emb = base_cru_1p.copy()
  base_cru_1p_emb['parafraseador'] = base_cru_1p_emb['texto'].progress_apply(lambda x:aug_ppdb.augment(x,num_gen))
  base_cru_1p_emb = base_cru_1p_emb.explode('parafraseador').reset_index(drop=True)
  exp_ppdb.append(base_cru_1p_emb)

  arquivo = '/content/drive/My Drive/Colab Notebooks/paper_asoc/perc_' +str(perc) + '/base_emotions_'+str(perc)+'_ppdb'+str(i+1)+'.csv'
  exp_ppdb[i].to_csv(arquivo, encoding='utf-8', index=False, sep=';', decimal = ',')
```
