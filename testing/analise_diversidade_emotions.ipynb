```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('sentence-transformers/paraphrase-albert-small-v2')
```

```python
dataset = [keyboard, eda, embed, bert, back_trans, deep_back, parrot, protaug,lambada, ppdb, ssmba]

for base in dataset:
  original_embedding = model.encode(base['texto'].astype('str'))
  base['texto_embedding'] = [original_embedding[i] for i in range(original_embedding.shape[0])]
  parafrase_embedding = model.encode(base['parafraseador'].astype('str'))
  base['parafraseador_embedding'] = [parafrase_embedding[i] for i in range(parafrase_embedding.shape[0])]
```

```python
for base in dataset:
  base['similiaridade_parafrase'] = base.progress_apply(lambda x:util.cos_sim(x['texto_embedding'], x['parafraseador_embedding']).tolist()[0][0], axis=1)
```
