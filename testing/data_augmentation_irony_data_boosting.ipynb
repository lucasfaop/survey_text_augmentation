```python
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer,PorterStemmer
from nltk.corpus import stopwords
import re
# lemmatizer = WordNetLemmatizer()
# stemmer = PorterStemmer() 

def preprocess(sentence):
  sentence=str(sentence)
  sentence = sentence.lower()
  sentence=sentence.replace('{html}',"") 
  cleanr = re.compile('<.*?>')
  cleantext = re.sub(cleanr, '', sentence)
  rem_url=re.sub(r'http\S+', '',cleantext)
  rem_num = re.sub('[0-9]+', '', rem_url)
  tokenizer = RegexpTokenizer(r'\w+')
  tokens = tokenizer.tokenize(rem_num)  
  filtered_words = [w for w in tokens if len(w) > 2 if not w in stopwords.words('english')]
  # stem_words=[stemmer.stem(w) for w in filtered_words]
  # lemma_words=[lemmatizer.lemmatize(w) for w in stem_words]
  return " ".join(filtered_words)


base_treino['cleanTexto']=base_treino['texto'].map(lambda s:preprocess(s)) 
```

```python
textos = base_treino['cleanTexto'].values
todos_textos = ' '.join(textos)
```

```python
from collections import Counter
from itertools import chain
from nltk.tokenize import sent_tokenize, word_tokenize

wordlist = list(chain(*[word_tokenize(s) for s in sent_tokenize(todos_textos)]))
contador_total = Counter(wordlist)
```

```python
nao_ironico = base_treino[base_treino['label'] == 0].copy()


textos_nao_ironicos = nao_ironico['cleanTexto'].values
todos_textos_nao_ironicos = ' '.join(textos_nao_ironicos)

wordlist_nao_ironicos = list(chain(*[word_tokenize(s) for s in sent_tokenize(todos_textos_nao_ironicos)]))
contador_total_nao_ironicos = Counter(wordlist_nao_ironicos)
```

```python
from sentence_transformers import SentenceTransformer, util
import math


def ganho(texto, lista_palavras, modelo):

  emb_texto = modelo.encode(texto , convert_to_tensor=True)
  emb_palavra = modelo.encode(lista_palavras, convert_to_tensor=True)
  cos_scores = util.cos_sim(emb_texto, emb_palavra)[0]

  soma = 0

  for scores in cos_scores:
    soma += math.log(1 +scores)
  
  return soma

```

# Treinamento Modelo Reforço

```python
import torch
import wandb
import time
import os
from tqdm import tqdm
import numpy as np
import pandas as pd
tqdm.pandas()

from transformers import GPT2Tokenizer
from transformers import AutoModelForSequenceClassification, AutoTokenizer

from trl.gpt2 import GPT2HeadWithValueModel, respond_to_batch
from trl.ppo import PPOTrainer
from trl.core import build_bert_batch_from_txt
```

```python
config = {
    "lm_name": "gpt2",
    "ref_lm_name": "gpt2",
    "cls_model_name": "lvwerra/bert-imdb",
    "tk_name": "gpt2",
    "steps": 2560,
    "batch_size": 64,
    "forward_batch_size": 64,
    "ppo_epochs": 10,   
    "txt_in_len": 2,
    "txt_out_len": 25,
    "lr": 1.41e-5,
    "init_kl_coef":0.2,
    "target": 6,
    "horizon":10000,
    "gamma":1,
    "lam":0.95,
    "cliprange": .2,
    "cliprange_value":.2,
    "vf_coef":.1, 
}
```

```python
gpt2_model_ironico = GPT2HeadWithValueModel.from_pretrained(config['lm_name'])
gpt2_model_nao_ironico = GPT2HeadWithValueModel.from_pretrained(config['lm_name'])
gpt2_model_ref = GPT2HeadWithValueModel.from_pretrained(config['ref_lm_name'])
gpt2_tokenizer = GPT2Tokenizer.from_pretrained(config['tk_name'])
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
_ = gpt2_model_ironico.to(device)
_ = gpt2_model_nao_ironico.to(device)
# _ = sentiment_model.to(device)
_ = gpt2_model_ref.to(device)
```

```python
df_ironico = base_treino[base_treino['label'] == 1].copy()
df_nao_ironico = base_treino[base_treino['label'] == 0].copy()
```

```python
df_ironico['tokens'] = df_ironico['texto'].progress_apply(lambda x: gpt2_tokenizer.encode(x, return_tensors="pt").to(device)[0, :config['txt_in_len']])
df_ironico['query'] = df_ironico['tokens'].progress_apply(lambda x: gpt2_tokenizer.decode(x))


df_nao_ironico['tokens'] = df_nao_ironico['texto'].progress_apply(lambda x: gpt2_tokenizer.encode(x, return_tensors="pt").to(device)[0, :config['txt_in_len']])
df_nao_ironico['query'] = df_nao_ironico['tokens'].progress_apply(lambda x: gpt2_tokenizer.decode(x))
```

```python
wandb.init(name='run-42', project='gpt2-data-boosting', config=config)
```

```python
wandb.watch(gpt2_model_nao_ironico, log='all')
```

```python
ppo_trainer = PPOTrainer(gpt2_model_nao_ironico, gpt2_model_ref, **config)
fbs = config['forward_batch_size']

for epoch in tqdm(range(int(np.ceil(config["steps"]/config['batch_size'])/20))):
    torch.cuda.empty_cache()
    logs = dict()
    game_data = dict()
    timing = dict()
    t0 = time.time()
    
    #### get a batch from the dataset
    df_batch = df_nao_ironico.sample(config['batch_size'])
    # print(df_batch.shape)
    game_data['query'] = df_batch['query'].tolist()
    query_tensors = torch.stack(df_batch['tokens'].tolist())
    
    #### get response from gpt2
    t = time.time()
    total_length = config['txt_in_len']+config['txt_out_len']
    response_tensors = []
    for i in range(int(config['batch_size']/fbs)):
        response  = respond_to_batch(gpt2_model_nao_ironico, query_tensors[i*fbs:(i+1)*fbs],
                                     txt_len=config['txt_out_len'])
        response_tensors.append(response)
    response_tensors = torch.cat(response_tensors)
    game_data['response'] = [gpt2_tokenizer.decode(response_tensors[i, :]) for i in range(config['batch_size'])]
    timing['time/get_response'] = time.time()-t

    # #### tokenize text for sentiment analysis
    # t = time.time()
    texts = game_data['response']
    # print(texts)
    # sentiment_inputs, attention_masks = build_bert_batch_from_txt(texts, sentiment_tokenizer, device)    
    # timing['time/build_input_sentiment'] = time.time()-t

    #### get rewards
    t = time.time()
    rewards = []
    for texto in texts:
        res = ganho(texto, top_ironicos , model_emb)
        # print(texts[i*fbs:(i+1)*fbs])
        # print(res)
        rewards.append(res)
    rewards = torch.tensor(rewards).to(device)
    # print(rewards)
    timing['time/get_sentiment_preds'] = time.time()-t

    #### Run PPO training 
    t = time.time()
    # print(query_tensors.shape)
    # print(response_tensors.shape)
    # print(rewards.shape)
    stats = ppo_trainer.step(query_tensors, response_tensors, rewards)
    timing['time/optimization'] = time.time()-t
     
    #### Log everything
    timing['time/epoch'] = time.time()-t0
    table_rows = [list(r) for r in zip(game_data['query'], game_data['response'], rewards.cpu().tolist())]
    logs.update({'game_log':wandb.Table(
        columns=['query', 'response', 'reward'],
        rows=table_rows)})
    logs.update(timing)
    logs.update(stats)
    logs['env/reward_mean'] = torch.mean(rewards).cpu().numpy()
    logs['env/reward_std'] = torch.std(rewards).cpu().numpy()
    logs['env/reward_dist'] = rewards.cpu().numpy()
    wandb.log(logs)
```
